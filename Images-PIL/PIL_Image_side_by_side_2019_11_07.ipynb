```python
def get_concat_h(im1, im2):
    #im1 = Image.open(im1)
    #im2 = Image.open(im2)
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst

def get_concat_v(im1, im2):
    #im1 = Image.open(im1)
    #im2 = Image.open(im2)
    dst = Image.new('RGB', (im1.width, im1.height + im2.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (0, im1.height))
    return dst

def get_concat_h2(im1, im2, im3):
    #im1 = Image.open(im1)
    #im2 = Image.open(im2)
    #im3 = Image.open(im3)
    HEIGHT = im1.height
    List = [im1.height,im2.height,im3.height]
    HEIGHT = max(List)
    dst = Image.new('RGB', (im1.width + im2.width + im3.width, HEIGHT))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    dst.paste(im3, (im1.width+im2.width, 0))
    return dst

print(sauce+im_vort)
im1 = sauce+im_vort
im2 = sauce+im_hilo


for i in range(len(files_hilo[0])):
    get_concat_h2(imgs_vort[i], imgs_hilo[i],imgs_pv[i]).save(f'/Users/chowdahead/Desktop/Weather_Blog/GFS/Ahhhh_{i}_.jpg')
    #get_concat_v(im1, im2).save('/Users/chowdahead/Desktop/pillow_concat_v.jpg')
```
